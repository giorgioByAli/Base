```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dosyayı yükleyin
from google.colab import files
uploaded = files.upload()

# Excel dosyasını pandas DataFrame olarak yükleme
data = pd.read_excel(list(uploaded.keys())[0])

# İlk 5 satırı görüntüleme
print(data.head())

# Boş değer kontrolü
print("Boş değerler:")
print(data.isnull().sum())

# Fiyat sütununun histogramı
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True, bins=30, color='blue')
plt.title('Fiyat Dağılımı', fontsize=16)
plt.xlabel('Fiyat', fontsize=12)
plt.ylabel('Frekans', fontsize=12)
plt.show()

# Oda sayısı ile fiyat arasındaki ilişki
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['room'], y=data['price'], hue=data['location'], palette='viridis')
plt.title('Oda Sayısı ile Fiyat Arasındaki İlişki', fontsize=16)
plt.xlabel('Oda Sayısı', fontsize=12)
plt.ylabel('Fiyat', fontsize=12)
plt.legend(title='Lokasyon')
plt.show()

# Yaş ile fiyat arasındaki ilişki
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['age'], y=data['price'], palette='Set3')
plt.title('Yaş ile Fiyat Dağılımı', fontsize=16)
plt.xlabel('Yaş', fontsize=12)
plt.ylabel('Fiyat', fontsize=12)
plt.show()

# Kat bilgisi (floor) dağılımı
plt.figure(figsize=(10, 6))
sns.countplot(x=data['floor'], palette='Set2')
plt.title('Kat Bilgisi Dağılımı', fontsize=16)
plt.xlabel('Kat', fontsize=12)
plt.ylabel('Frekans', fontsize=12)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Define the URL of the website to scrape
url = "https://www.sahibinden.com/satilik-daire/antalya"

# Headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Function to scrape data from a single page
def scrape_page(url):
    try:
        # Send a GET request to the website
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an error for bad status codes
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve the webpage: {e}")
        return None

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all listings (adjust the class name based on the website's structure)
    listings = soup.find_all("tr", class_="searchResultsItem")  # Example class name

    # Initialize a list to store the data
    data = []

    # Loop through each listing and extract details
    for listing in listings:
        try:
            # Extract title
            title = listing.find("td", class_="searchResultsTitleValue").text.strip()

            # Extract price
            price = listing.find("td", class_="searchResultsPriceValue").text.strip()

            # Extract location
            location = listing.find("td", class_="searchResultsLocationValue").text.strip()

            # Extract size and room info (if available)
            details = listing.find("td", class_="searchResultsAttributeValue").text.strip()

            # Append the data to the list
            data.append([title, price, location, details])
        except AttributeError:
            # Skip listings with missing data
            continue

    return data

# Function to scrape multiple pages
def scrape_multiple_pages(base_url, num_pages):
    all_data = []
    for page in range(1, num_pages + 1):
        print(f"Scraping page {page}...")
        page_url = f"{base_url}?pagingOffset={(page - 1) * 20}"  # Adjust for pagination
        page_data = scrape_page(page_url)
        if page_data:
            all_data.extend(page_data)
        time.sleep(2)  # Add a delay to avoid overwhelming the server
    return all_data

# Scrape data from multiple pages
num_pages = 5  # Number of pages to scrape
all_data = scrape_multiple_pages(url, num_pages)

# Check if data is available before saving
if not all_data:
    print("No data found to save.")
else:
    # Convert the data into a pandas DataFrame
    columns = ["Title", "Price", "Location", "Details"]
    df = pd.DataFrame(all_data, columns=columns)

    # Save the data to an Excel file
    df.to_excel("antalya_house_listings.xlsx", index=False)
    print("Data saved to antalya_house_listings.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

# Define the URL of the website to scrape
url = "https://www.sahibinden.com/satilik/antalya"

# Headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Referer": "https://www.google.com/",
    "Connection": "keep-alive"
}

# Function to scrape data from a single page
def scrape_page(url):
    try:
        # Send a GET request to the website
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an error for bad status codes
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve the webpage: {e}")
        return None

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all listings (adjust the class name based on the website's structure)
    listings = soup.find_all("tr", class_="searchResultsItem")  # Example class name

    # Initialize a list to store the data
    data = []

    # Loop through each listing and extract details
    for listing in listings:
        try:
            # Extract title
            title = listing.find("td", class_="searchResultsTitleValue").text.strip()

            # Extract price
            price = listing.find("td", class_="searchResultsPriceValue").text.strip()

            # Extract location
            location = listing.find("td", class_="searchResultsLocationValue").text.strip()

            # Extract size and room info (if available)
            details = listing.find("td", class_="searchResultsAttributeValue").text.strip()

            # Append the data to the list
            data.append([title, price, location, details])
        except AttributeError:
            # Skip listings with missing data
            continue

    return data

# Function to scrape multiple pages
def scrape_multiple_pages(base_url, num_pages):
    all_data = []
    for page in range(1, num_pages + 1):
        print(f"Scraping page {page}...")
        page_url = f"{base_url}?pagingOffset={(page - 1) * 20}"  # Adjust for pagination
        page_data = scrape_page(page_url)
        if page_data:
            all_data.extend(page_data)
        time.sleep(random.uniform(1, 3))  # Random delay to avoid blocking
    return all_data

# Scrape data from multiple pages
num_pages = 5  # Number of pages to scrape
all_data = scrape_multiple_pages(url, num_pages)

# Check if data is available before saving
if not all_data:
    print("No data found to save.")
else:
    # Convert the data into a pandas DataFrame
    columns = ["Title", "Price", "Location", "Details"]
    df = pd.DataFrame(all_data, columns=columns)

    # Save the data to an Excel file
    df.to_excel("antalya_house_listings.xlsx", index=False)
    print("Data saved to antalya_house_listings.xlsx")
```
